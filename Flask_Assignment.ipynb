### Question 1: What is a RESTful API?

**Answer:**
A RESTful API is an architectural style designed for web applications to talk to each other over HTTP. It's built around the concept of "resources"—which basically means any piece of data (like a user account, a product listing, or an image) gets its own unique URL.

To count as truly RESTful, the system has to follow a few core rules:
* Statelessness: This is a big one. The server doesn't remember past requests. Every single request from the client has to bring along all the context and data needed to complete it.
* Client-Server Separation: The frontend (client) and backend (server) are totally decoupled. You can change the database structure without breaking how the mobile app looks.
* Cacheability: Responses say whether they can be cached or not, which keeps the network from getting bogged down with repeated requests.
* Uniform Interface: Everything uses standard, predictable HTTP methods so developers don't have to guess how to interact with the API.

### Question 2: What is Flask, and why is it popular for building APIs?

**Answer:**
Flask is a micro web framework written in Python. It gets called a "micro" framework because it doesn't force you into using specific tools, database wrappers, or complex folder structures right out of the box. It gives you the absolute bare essentials to handle web requests and leaves the rest up to you.

It's incredibly popular for building APIs for a few reasons:
* Minimal boilerplate: You don't have to write dozens of setup files. You can literally get a working API endpoint up and running in under ten lines of code.
* Total control: Unlike heavier frameworks like Django, Flask doesn't make assumptions for you. You pick your own database tools, your own validation libraries, and your own authentication methods.
* Lightweight and quick: Because it doesn't carry unnecessary weight, it's very fast and fits perfectly into modern microservices and simple backend setups.
* Great extension system: If you do need extra features later on, you can just plug in community-managed tools like Flask-SQLAlchemy or Flask-JWT.

### Question 3: What are HTTP methods used in RESTful APIs?

**Answer:**
HTTP methods tell the server exactly what action we want to perform on a resource. The most common ones correspond directly to standard CRUD operations:

| HTTP Method | Main Purpose | What it maps to |
| :--- | :--- | :--- |
| **GET** | Fetches data from the server. It should never alter the database state. | Read |
| **POST** | Sends new data to the server to create a brand new record. | Create |
| **PUT** | Replaces an entire existing resource with a new version. | Update (Full Replacement) |
| **PATCH** | Tweaks specific fields of an existing resource instead of replacing the whole thing. | Update (Partial Modify) |
| **DELETE** | Wipes out a specific resource from the server completely. | Delete |

**Answer:**
The `@app.route()` decorator acts as a traffic controller for your code. It binds a specific URL path to a Python function. When a user or a frontend app hits that URL, Flask looks up the decorator, intercepts the request, runs the function tied to it, and sends the function's return value back to the user.

```python
from flask import Flask
app = Flask(__name__)

@app.route('/dashboard-data')
def fetch_dashboard():
    return "This string is sent back whenever someone hits the /dashboard-data URL."
```

### Question 5: What is the role of Flask-SQLAlchemy?

**Answer:**
Flask-SQLAlchemy is a wrapper extension that links Flask up with SQLAlchemy, which is an Object-Relational Mapper (ORM). 

Instead of making you write messy, raw SQL queries inside your Python strings, it lets you interact with database tables using standard Python classes and methods. You can query models, filter rows, and insert data completely through object-oriented Python, while the extension handles opening connections, managing sessions, and preventing SQL injection attacks in the background.

### Question 7: How do you return JSON responses in Flask?

**Answer:**
In Flask, you can return a native Python dictionary or list from your route function, and Flask will automatically handle turning it into a JSON string. However, explicitly using the built-in `jsonify()` helper function is still best practice because it automatically injects the proper `Content-Type: application/json` header into the HTTP response so frontend clients process it correctly.

```python
from flask import jsonify

@app.route('/api/status')
def check_system_status():
    output_payload = {
        "server_healthy": True,
        "active_connections": 42
    }
    return jsonify(output_payload), 200
```

### Question 8: How do you handle POST requests in Flask?

**Answer:**
Flask routes accept `GET` requests by default. If you want to accept data inputs via a `POST` request, you have to pass a `methods=['POST']` list to the route decorator. Then, inside the function, you use the global `request` object to extract the incoming body data (usually using `request.get_json()`).

```python
from flask import request, jsonify

@app.route('/api/add-item', methods=['POST'])
def process_new_item():
    incoming_payload = request.get_json()
    product_title = incoming_payload.get('title', 'Untitled')
    return jsonify({
        "status": "created",
        "saved_item": product_title
    }), 201
```

### Question 9: How do you handle errors in Flask (e.g., 404)?

**Answer:**
You can clean up broken endpoints or missing resources using the `@app.errorhandler()` decorator. It lets you capture specific HTTP status errors (like 404 for missing pages or 500 for internal server crashes) and override Flask's default HTML errors with custom, clean JSON payloads.

```python
from flask import jsonify

@app.errorhandler(404)
def handle_missing_resource(error):
    return jsonify({
        "success": False,
        "error_code": 404,
        "message": "The requested URL or object could not be found."
    }), 404
```

### Question 10: How do you structure a Flask app using Blueprints?

**Answer:**
When an application grows, keeping all your routes in a single file makes the code a nightmare to maintain. **Blueprints** solve this by letting you break your application into separate, independent modules or components (like separate files for authentication routes, profile routes, or payment routes).

You initialize a Blueprint inside a component file, write out your view functions just like normal, and then import and register that blueprint object inside your main entry point file (`app.py`). This keeps everything organized, clean, and isolated.

```python
from flask import Flask, Blueprint

catalog_bp = Blueprint('catalog', __name__)

@catalog_bp.route('/items')
def view_items():
    return "List of catalog products"

main_app = Flask(__name__)\main_app.register_blueprint(catalog_bp, url_prefix='/products')
```
